```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('census_data.csv')
df
```

```python
# stateicp: 54
# county: 370

totpop1820 = df[(df['stateicp'] == 54) & (df['county'] == 370) & (df['year'] == 1820)]['citypop'].sum()
totpop1830 = df[(df['stateicp'] == 54) & (df['county'] == 370) & (df['year'] == 1830)]['citypop'].sum()
totpop1840 = df[(df['stateicp'] == 54) & (df['county'] == 370) & (df['year'] == 1840)]['citypop'].sum()
```

```python
totpop1820 = round(totpop1820 / 1000000, 2)
totpop1830 = round(totpop1830 / 1000000, 2)
totpop1840 = round(totpop1840 / 1000000, 2)
```

```python
x = ['1820', '1830', '1840']
y = [totpop1820, totpop1830, totpop1840]
```

```python
plt.plot(x, y)
plt.xlabel('Year')  # add X-axis label
plt.ylabel('Total Population (Millions)')  # add Y-axis label
plt.title('Population Change in Davidson County, Tennessee between 1820 and 1840')
plt.show()
```
